```python
import pandas as pd

data_path = "D:/Users/Documents/Jupyter/BreastCancerClassification/data/breast-cancer.data"

BreastCancer = pd.read_csv(data_path)

BreastCancer.columns = ["Class", "Age", "menopause", "tumor-size", "inv-nodes", "node-caps", "deg-malig", "breast", "breast-quad", "irradiat"]

BreastCancer.head()
```

```python
BreastCancer.info()

BreastCancer.isnull().sum()

BreastCancer["Class"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = BreastCancer.select_dtypes(include = ["object"]).columns.tolist()

encoder = OneHotEncoder(sparse_output = False)

one_hot_encoded = encoder.fit_transform(BreastCancer[categorical_columns])

one_hot_BreastCancer = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))

BreastCancer_encoded = pd.concat([BreastCancer, one_hot_BreastCancer], axis = 1)

BreastCancer_encoded = BreastCancer_encoded.drop(categorical_columns, axis=1)

print(BreastCancer_encoded.head(5))
```
